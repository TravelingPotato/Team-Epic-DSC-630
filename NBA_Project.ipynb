Corr Heatmap of all features

```python
import numpy as np

corr = nba.corr()

f, ax = plt.subplots(figsize=(11, 9))

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, cmap=cmap, mask=mask, center=0,
           square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

nba_subset2 uses the features from feature selection doc with a score higher than 2 minus "player name" and "team"

```python
nba_subset2 = nba[["Age", "Player_Efficiency_Rating", "Minutes_Played", "Blocks", "Free_Throws_Attempted", "Turnovers_Pct", "Two_Pointers_Attempted", "Steals", "Points", "Assists_Pct", "Games_Started","Games_Played", "Total_Rebounds", "Field_Goals_Attempted", "Field_Goals", "Defensive_Box_Plus_Minus", "Defensive_Rebound_Pct", "Usage_Pct", "Free-Throws"]]
```

```python
corr = nba_subset2.corr()

f, ax = plt.subplots(figsize=(11, 9))

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, cmap=cmap, mask=mask, center=0,
           square=True, linewidths=.5,cbar_kws={"shrink": .5})
```

```python
nba_subset3 = nba_subset2.drop("Minutes_Played", axis=1)
nba_subset3 = nba_subset3.drop("Two_Pointers_Attempted", axis=1)
nba_subset3 = nba_subset3.drop("Points", axis=1)
nba_subset3 = nba_subset3.drop("Free_Throws_Attempted", axis=1)
nba_subset3 = nba_subset3.drop("Field_Goals_Attempted", axis=1)
```

```python
corr = nba_subset3.corr()

f, ax = plt.subplots(figsize=(11, 9))

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, cmap=cmap, mask=mask, center=0,
           square=True, linewidths=.5,
            annot=True,cbar_kws={"shrink": .5})
```

```python
nba_used = nba_subset3
nba_used['Salary'] = nba["Salary"]
nba_used = nba_used.dropna()
nba_used.shape
```

```python
from sklearn.linear_model import LinearRegression
x = nba_used.drop('Salary', axis=1)
y = nba_used['Salary']

model = LinearRegression().fit(x,y)

r_sq = model.score(x,y)

r_sq
```
